```python
import matplotlib.pyplot as plt #Primeiro tipo de gráfico

import plotly.offline as py   #Segundo tipo de gráfico
import plotly.graph_objs as go

from ipywidgets import interact #Pacote de interatividade
import ipywidgets as widgets

import math as m #pacote de cálculos

G = 6.67408
M = 1.989
delta_tempo = 0.0001
```

```python
def movimento(r_0, v_x_0,v_y_0,tempo):
        x = []
        y = []
        x.append(r_0)
        y.append(0)
        v_x = v_x_0
        v_y = v_y_0
        delta_tempo_dias = (delta_tempo*pow(10,7))/86400
        n = int(tempo/(delta_tempo_dias))
        for i in range(1,n):
                x.append(x[i-1] + v_x*delta_tempo)
                y.append(y[i-1] + v_y*delta_tempo)
                
                r_2 = pow(x[i-1],2)+pow(y[i-1],2)
                a = -(G*M)/r_2
                a_x = (x[i-1]/m.sqrt(r_2))*a
                a_y = (y[i-1]/m.sqrt(r_2))*a
                v_x = v_x + a_x*delta_tempo
                v_y = v_y + a_y*delta_tempo
        return x,y
```

```python
def plotar_movimento(movimento):
        plt.title('Trajetória em torno do Sol',fontsize = 20)
        plt.plot(movimento[0],movimento[1],label = 'Trajetória do planeta')
        plt.scatter(0,0,color= 'orange', label = 'Sol')
        plt.legend(loc=[1.1,0.81],fontsize = 15)
        plt.xlabel('X (10^11) metros',fontsize = 15)
        plt.ylabel('Y (10^11) metros',fontsize = 15)
        plt.rcParams['figure.figsize'] = (5,5)
        plt.show() 
```
